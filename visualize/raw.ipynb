```python
from helper import DataPreprocessing
from ipywidgets import interact, IntSlider
import matplotlib.pyplot as plt
import numpy as np
```

```python
kind = "test/full_test"
#load data test
data_test = DataPreprocessing("test")
data_full = DataPreprocessing("full")
for bearing in [13,14,15,16,17,23,24,25,26,27,33]:
    
    array_test = data_test.load_array(bearing)
    N_test = array_test.shape[1]
    last_cycle = data_test.last_cycle(bearing)
    time_stamp_test = np.linspace(0,last_cycle,N_test)
    rul = data_test.rul(bearing)

    hor_acc_test = array_test[0]
    ver_acc_test = array_test[1]
    
    array_full = data_full.load_array(bearing)                         
    N_full = array_full.shape[1]
    eol = data_test.eol(bearing)
    time_stamp_full = np.linspace(0,eol,N_full)
    
    hor_acc_full = array_full[0]
    ver_acc_full = array_full[1]

    #visualize
    fig, (ax1, ax2) = plt.subplots(nrows = 1, ncols = 2,figsize = (15,5))
    fig.suptitle("{} data, bearing {}, EOL {} s, RUL {} s".format(kind,bearing,eol,rul))

    #axes 1:
    ax1.plot(time_stamp_full,hor_acc_full,c="green")
    ax1.plot(time_stamp_test,hor_acc_test,c="blue")
    ax1.legend(["horizontal test","horizontal full"],loc="upper left")
    ax1.set_ylabel("Vibration in g")
    ax1.set_xlabel("Timestamp in s")
    ax1.set_title("Horizontal vibration")

    #axes 2:
    ax2.plot(time_stamp_full,ver_acc_full,c="green")
    ax2.plot(time_stamp_test,ver_acc_test,c="red")         
    ax2.legend(["vertical test","vertical full"],loc="upper left")
    ax2.set_ylabel("Vibration in g")
    ax2.set_xlabel("Timestamp in s")
    ax2.set_title("Vertical vibration")

    print("Done")
```

```python
#load data
kind = "train"
data = DataPreprocessing(kind)

for bearing in [11,12,21,22,31,32]:
    
    #load data
    N = data.n_samples(kind,bearing)
    eol = data.eol(bearing)
    time_stamp = np.linspace(0,eol,N)
    array = data.load_array(bearing)

    hor_acc = array[0]
    ver_acc = array[1]

    #visualize
    fig, (ax1, ax2) = plt.subplots(nrows = 1, ncols = 2,figsize = (15,5))
    fig.suptitle("{} data, bearing {}, EOL {} s".format(kind,bearing,eol))

    #axes 1:
    ax1.plot(time_stamp,hor_acc)
    ax1.legend(["horizontal"],loc="upper left")
    ax1.set_ylabel("Vibration in g")
    ax1.set_xlabel("Timestamp in s")
    ax1.set_title("Horizontal vibration")

    #axes 2:
    ax2.plot(time_stamp,ver_acc,c = "red")
    ax2.legend(["vertical"],loc="upper left")
    ax2.set_ylabel("Vibration in g")
    ax2.set_xlabel("Timestamp in s")
    ax2.set_title("Vertical vibration")

    print("Done")
```
